```python
import pyzx as zx


def do_funny(thing):
    zx.draw(thing)
    print("gates: ", len(thing.gates))
    print("depth: ", thing.depth())
    print("t-gates: ", thing.tcount())
    print("two-qubit-gates: ", thing.twoqubitcount())
    
qubit_amount = 5
gate_count = 80
#Generate random circuit of Clifford gates
cliffordTGraph = zx.generate.cliffordT(qubit_amount, gate_count)
#If running in Jupyter, draw the circuit
extracted_circuit = zx.extract_circuit(cliffordTGraph.copy())
copy = cliffordTGraph.copy()

do_funny(extracted_circuit)

zx.clifford_simp(copy)
extracted_circuit = zx.extract_circuit(copy)
do_funny(extracted_circuit)


```

```python
qasm = ""
with open("circuits/small/adder_8.qasm", 'r') as file:
    file_contents = file.read()
    qasm = file_contents   
original = zx.Circuit.from_qasm(qasm)
original_copy = original.copy()

do_funny(original)
qasm_graph = original.to_graph()

zx.clifford_simp(qasm_graph)
extracted_circuit = zx.extract_circuit(qasm_graph)
do_funny(extracted_circuit)

new_circuit = zx.optimize.full_optimize(original_copy)
do_funny(new_circuit)
```
